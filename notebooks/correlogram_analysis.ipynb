```python
# Imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = r'C:\Users\vitorlvg\Documents\Python_Scripts\indexes\data\data\df_correlogram.xlsx'

df_lag1 = pd.read_excel(path, sheet_name='lag1', usecols='A:AL', engine='openpyxl', )
df_lag2 = pd.read_excel(path, sheet_name='lag2', usecols='A:AL', engine='openpyxl')
```

```python
df_lag1.drop(columns=['time_sin', 'time_tel_lag1'], inplace=True)
df_lag2.drop(columns=['time_sin', 'time_tel_lag2'], inplace=True)
```

```python
# Correlogram
corr_lag1 = df_lag1.corr()

# mask
mask = np.triu(np.ones_like(corr_lag1, dtype=bool))

# Plotting
plt.rcParams["figure.figsize"] = (28, 15)
sns.heatmap(corr_lag1, fmt=".2f", mask=mask, annot=True, cmap='RdBu', center=0, vmin=-1, vmax=1,
            cbar_kws={"shrink": 0.8}, annot_kws={"size": 14})
plt.title('Correlogram lag=1 - Teleconnections, SACZ, and Blocking for Brazil', fontsize=20)
plt.xticks(rotation=45, ha='right', fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.savefig(r"correlogram_lag1.png", dpi=300)
plt.show()
```

```python
# Correlogram
corr_lag2 = df_lag2.corr()

# mask
mask = np.triu(np.ones_like(corr_lag2, dtype=bool))

# Plotting
plt.rcParams["figure.figsize"] = (28, 15)
sns.heatmap(corr_lag2, fmt=".2f", mask=mask, annot=True, cmap='RdBu', center=0, vmin=-1, vmax=1,
            cbar_kws={"shrink": 0.8}, annot_kws={"size": 14})
plt.title('Correlogram lag=2 - Teleconnections, SACZ, and Blocking for Brazil', fontsize=20)
plt.xticks(rotation=45, ha='right', fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.savefig(r"correlogram_lag2.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pth = r'C:\Users\vitorlvg\Documents\Python_Scripts\indexes\merged_indices_vf3_correlogram.csv'
df = pd.read_csv(pth, sep=';')
df['time'] = pd.to_datetime(df['time'])
df.set_index('time', inplace=True)

df_filtered = df[df.index.day.isin([10, 11, 12, 1, 2, 3, 4])]
df_filtered.to_csv('indexes_final_version.csv')

# heatmap
corr_df = df_filtered.corr()

# mask
mask = np.triu(np.ones_like(corr_df, dtype=bool))

# Plotting
plt.rcParams["figure.figsize"] = (28, 15)
sns.heatmap(corr_df, fmt=".2f", mask=mask, annot=True, cmap='RdBu', center=0, vmin=-1, vmax=1,
            cbar_kws={"shrink": 0.8}, annot_kws={"size": 14})
plt.title('Correlation Matrix - Teleconnections, SACZ, and Blocking for Brazil', fontsize=20)
plt.xticks(rotation=45, ha='right', fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.savefig(r"correlation_matrix_tele_sacz_blocking.png", dpi=300)
plt.show()
```
